```python
def validate_taxonomy_row(row):
    
    validation = {}
    warnings = []

    # Validación de GENUS
    validation['GENUS_VALID'] = 1 if pd.notna(row['GENUS']) else 0
    
    # Si GENUS es válido, continuamos con el resto
    if validation['GENUS_VALID'] == 1:
              
        if pd.notna(row['SUBTAUTHOR']):
            validation['SUBTAXA_VALID'] = 1 if pd.notna(row['SUBTAXA']) & pd.notna(row['SPECIES']) else 0
           
        else:
             warnings.append('Subtaxa authority description is missing')
             validation['SUBTAXA_VALID'] = 0.5 if pd.notna(row['SUBTAXA']) & pd.notna(row['SPECIES']) else 0

        if pd.notna(row['SPAUTHOR']):
            validation['SPECIES_VALID'] = 1 if pd.notna(row['SPECIES']) else 0
        else:
             warnings.append('Species authority description is missing')
             validation['SPECIES_VALID'] = 0.5 if pd.notna(row['SPECIES']) else 0

        if pd.notna(row['SPECIES']):

            Spp = bool(re.search(r'\b(sp|spp|sp\.)\b', row['SPECIES'], re.IGNORECASE))

            if Spp: 
                validation['SPECIES_VALID'] = 0
                warnings.append('Species name like Sp, spp, sp by itself is not is not descriptive')
        else:
             warnings.append('Species name is missing')

        if pd.notna(row['SUBTAXA']):
        
            Subspp = bool(re.search(r'f\.|var\.|subsp\.', row['SUBTAXA'], re.IGNORECASE)) 

            if Subspp == False:
                validation['SUBTAXA_VALID'] = validation['SUBTAXA_VALID'] - 0.25
                warnings.append('var. susbsp, f. etc.. is missing as first word of the field. (This can create taxonomic issues)')

        else:
            warnings.append('Subtaxa name is missing')


    else:

         warnings.append('Genus description is missing')
         validation['SUBTAXA_VALID'] = 0
         validation['SPECIES_VALID'] = 0



    # Cálculo del puntaje total basado en la suma de los validadores
    validation['TAXA_SCORE_VALUE'] = sum(validation.values())
    validation['WARNINGS'] = '; '.join(warnings) if warnings else None
    
    return pd.Series(validation)
```

```python
# Función para aplicar la validación a todo el DataFrame
def validate_taxonomy_df(df):
    validation_df = df.apply(validate_taxonomy_row, axis=1)
    return pd.concat([df, validation_df], axis=1)
```

```python
def traffic_ligth(df):

    df['TAXA_SCORE_CATEGORY'] = ['Green' if value > 2 and value <= 3 else 'Yellow' if value > 1 and value <= 2 else 'Red' for value in df['TAXA_SCORE_VALUE']]

    return df
```

```python
def check_taxonomy_suggestions(row, grin):

    row['GRIN_TAXON_ID'] = None
    validation = {}
    if pd.notna(row['GRIN_TAXON_ID']):
        data = grin.loc[grin['taxonomy_species_id'] == row['GRIN_TAXON_ID']]
        
        if not data.empty:
            # Extract the first matching row as a Series
            data = data.iloc[0]
            
            # Safely extract each component, replacing NaN with empty strings
            genus = str(row.get('GENUS', '')).strip()
            species_name = str(data.get('species_name', '')).strip()
            species_authority = str(data.get('species_authority', '')).strip()
            subspecies_name = str(data.get('subspecies_name', '')).strip()
            subspecies_authority = str(data.get('subspecies_authority', '')).strip()
            variety_name = str(data.get('variety_name', '')).strip()
            variety_authority = str(data.get('variety_authority', '')).strip()
            forma_name = str(data.get('forma_name', '')).strip()
            forma_authority = str(data.get('forma_authority', '')).strip()
            
            # Concatenate the parts
            suggested_taxa = (
                f"{genus} {species_name} {species_authority} "
                f"subsp. {subspecies_name} {subspecies_authority} "
                f"var. {variety_name} {variety_authority} "
                f"f. {forma_name} {forma_authority}"
            )


        validation['suggested_taxa']  = re.sub(r'subsp\. \\N|var\. \\N|f\. \\N', '', suggested_taxa)
        validation['suggested_taxa'] = re.sub(r'\\N',  '', validation['suggested_taxa'])
        validation['suggested_taxa'] = re.sub(r'\\',  '', validation['suggested_taxa'])
        
        validation['suggested_grin_id'] =  row['GRIN_TAXON_ID']
   
    else:
        if pd.notna(row['SPECIES']) & pd.notna(row['SUBTAXA']):
                   
          data = grin.loc[(grin['name'] == row['SPECIES'] + ' ' + row['SUBTAXA'])]
         # data = data[(data['subspecies_name'].str.contains(row['SUBTAXA'], case=False, na=False)) | 
         #             (data['variety_name'].str.contains(row['SUBTAXA'], case=False, na=False)) | 
         #             (data['subvariety_name'].str.contains(row['SUBTAXA'], case=False, na=False)) |
         #             (data['forma_name'].str.contains(row['SUBTAXA'], case=False, na=False))]
          validation['suggested_grin_id'] =  data.get('current_taxonomy_species_id', '')

          # Safely extract each component, replacing NaN with empty strings
          genus = str(row.get('GENUS', '')).strip()
          species_name = str(data.get('species_name', '')).strip()
          species_authority = str(data.get('species_authority', '')).strip()
          subspecies_name = str(data.get('subspecies_name', '')).strip()
          subspecies_authority = str(data.get('subspecies_authority', '')).strip()
          variety_name = str(data.get('variety_name', '')).strip()
          variety_authority = str(data.get('variety_authority', '')).strip()
          forma_name = str(data.get('forma_name', '')).strip()
          forma_authority = str(data.get('forma_authority', '')).strip()
            

          validation['suggested_taxa'] =  (
                f"{genus} {species_name} {species_authority} "
                f"subsp. {subspecies_name} {subspecies_authority} "
                f"var. {variety_name} {variety_authority} "
                f"f. {forma_name} {forma_authority}"
            )

          validation['suggested_taxa']  = re.sub(r'subsp\. \\N|var\. \\N|f\. \\N', '', validation['suggested_taxa'])
          validation['suggested_taxa'] = re.sub(r'\\N',  '', validation['suggested_taxa'])
          validation['suggested_taxa'] = re.sub(r'\\',  '', validation['suggested_taxa'])

     
        else: 
            if  pd.notna(row['SPECIES']):

                data = grin.loc[(grin['name'] == row['SPECIES'])]           
                validation['suggested_grin_id'] =  data.get('current_taxonomy_species_id', '')

                # Safely extract each component, replacing NaN with empty strings
                genus = str(row.get('GENUS', '')).strip()
                species_name = str(data.get('species_name', '')).strip()
                species_authority = str(data.get('species_authority', '')).strip()
                subspecies_name = str(data.get('subspecies_name', '')).strip()
                subspecies_authority = str(data.get('subspecies_authority', '')).strip()
                variety_name = str(data.get('variety_name', '')).strip()
                variety_authority = str(data.get('variety_authority', '')).strip()
                forma_name = str(data.get('forma_name', '')).strip()
                forma_authority = str(data.get('forma_authority', '')).strip()
                                  
                validation['suggested_taxa'] =  (
                           f"{genus} {species_name} {species_authority} "
                            f"subsp. {subspecies_name} {subspecies_authority} "
                            f"var. {variety_name} {variety_authority} "
                            f"f. {forma_name} {forma_authority}"
                        )
            else: 

                data = grin.loc[(grin['name']  ==  row['SUBTAXA'])]               
                #data = grin.copy()
                #data = data[(data['subspecies_name'].str.contains(row['SUBTAXA'], case=False, na=False)) | 
                #           (data['variety_name'].str.contains(row['SUBTAXA'], case=False, na=False)) | 
                #            (data['subvariety_name'].str.contains(row['SUBTAXA'], case=False, na=False)) |
                #            (data['forma_name'].str.contains(row['SUBTAXA'], case=False, na=False))]

                validation['suggested_grin_id'] =  data.get('current_taxonomy_species_id', '')
                        
                validation['suggested_taxa'] =  (
                            f"{genus} {species_name} {species_authority} "
                            f"subsp. {subspecies_name} {subspecies_authority} "
                            f"var. {variety_name} {variety_authority} "
                            f"f. {forma_name} {forma_authority}"
                        )
                        
    # Check if suggested_grin_id is a Series
    if isinstance(validation['suggested_grin_id'], pd.Series):
        if validation['suggested_grin_id'].notna().sum() > 1:  # Check for multiple non-null values
            # Concatenate non-null values into a single string
            validation['suggested_grin_id'] = ', '.join(
                validation['suggested_grin_id'].dropna().astype(str)
            )
        else:
            # Check for a single non-null value
            single_value = validation['suggested_grin_id'].dropna()
            if not single_value.empty:  # Check if there's a non-null value
                validation['suggested_grin_id'] = int(single_value.iloc[0])
            else:
                validation['suggested_grin_id'] = None  # Handle case where all values are NaN
    else:
        # If it's not a Series, it might be a single value (like float or int)
        if pd.notna(validation['suggested_grin_id']):
            validation['suggested_grin_id'] = int(validation['suggested_grin_id'])
        else:
            validation['suggested_grin_id'] = None  # Handle NaN


    return pd.Series(validation)

    #include join for multiple taxa id 
    # include the case taxa species id NA
```
